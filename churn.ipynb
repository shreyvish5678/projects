```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("Customer_Data_Telecom.csv")
dataset.drop("customerID", axis=1, inplace=True)
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
binary_dict = {'Yes': 1, 'No': 0, 'No phone service': 0, 'No internet service': 0}
binary_columns = ['Partner', 'Dependents', 'PhoneService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn', 'MultipleLines']
dataset[binary_columns] = dataset[binary_columns].replace(binary_dict)
contract_dict = {'Month-to-month': 1, 'One year': 2, 'Two year': 3}
dataset['Contract'] = dataset['Contract'].map(contract_dict)
gender_dict = {'Male': 1, 'Female': 0}
dataset['gender'] = dataset['gender'].map(gender_dict)
dataset.head(10)
```

```python
one_hot_columns = ['InternetService', 'PaymentMethod'] 
dataset = pd.get_dummies(data=dataset, columns=one_hot_columns, drop_first=False)
dataset = dataset.drop(['InternetService_No', 'PaymentMethod_Mailed check'], axis=1)
dataset.head()
```

```python
X = dataset.drop(['Churn', 'TotalCharges'], axis=1)
y = dataset['Churn']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
print(accuracy_score(y_pred, y_test))
```
